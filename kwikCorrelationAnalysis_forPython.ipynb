```python
import pandas as pd
df = pd.read_csv('NHP.csv')
df
```

```python
import statsmodels.api as sm
import numpy as np
import pandas as pd
import openpyxl
array = []
for col in df.columns:
    x = df[col]
    x1 = sm.add_constant(x, prepend = True)
    y = df['Exterior2nd']
    b = np.corrcoef(y,x)
    c = b[0,1].round(2)
    z = sm.OLS(y, x1).fit()
    a = z.pvalues[1]
    array.append([col, a, c])
output = pd.DataFrame(array, columns=['Category','P-value','Correlation'])
output.sort_values(by = ['Correlation'])
```
